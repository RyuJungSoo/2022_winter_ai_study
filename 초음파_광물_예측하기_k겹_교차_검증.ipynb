```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold

import numpy
import pandas as pd
import tensorflow as tf
```

```python
# seed 값 설정
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(seed)
```

```python
# 데이터 입력
df = pd.read_csv('../dataset/sonar.csv', header=None)

dataset = df.values
X = dataset[:,0:60].astype(numpy.float)
Y_obj = dataset[:,60]
```

```python
# 문자열 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
# 10개의 파일로 쪼갬
n_fold = 10
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed) # shuffle 하면 index를 섞는다
```

```python
# 빈 accuracy 배열
accuracy = []
```

```python
# 모델의 설정, 컴파일, 실행
for train, test in skf.split(X,Y):
  model = Sequential()
  model.add(Dense(24, input_dim=60, activation='relu'))
  model.add(Dense(10,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])  # 오차 함수로는 평균 제곱 오차 함수를 사용
  # 모델 실행
  model.fit(X[train], Y[train], epochs=130, batch_size=5)
  k_accuracy = "%.4f" % (model.evaluate(X[test], Y[test])[1])
  accuracy.append(k_accuracy)
```

```python
# 결과 출력
print("\n %.f fold accuracy:" % n_fold, accuracy)
```
